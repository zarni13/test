```python
import numpy as np
import matplotlib.pyplot as plt
N = 10000
h = 0.01
t_0 = 0
x_0 = 0
t = np.zeros(N+1)
x = np.zeros(N+1)
t[0] = t_0
x[0] = x_0
t_old = t_0
x_old = x_0
for n in range (N):
    x_new = x_old + h*(np.cos(x_old)+np.sin(t_old))
    t[n+1] = t_old + h
    x[n+1] = x_new
    t_old = t_old + h
    x_old = x_new
print(r'x_N= %f'% x_old)
plt.figure(figsize=(16,8))
plt.plot(t,x)
plt.ylabel("x(t)")
plt.xlabel("t")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_Euler(y,h,f):
    next_y = y + h*f(y)
    return next_y
```

```python
def full_Euler(h,f, start_t=0,end_t=1,y_0=1):
    N = int((end_t - start_t)/h)
    t_list=np.linspace(start_t,end_t,N+1)
    y_list=np.zeros(N+1)
    y_list[0]=y_0
    for i in range(0,N):
        y_list[i+1]=step_Euler(y_list[i],h,f)
        return y_list,t_list
```

```python
def g(y):
    growth_rate=np.log(2)*y
    return growth_rate
y_0 = 1  # Initial population size, i.e. a single bacteria
h = 0.01 # Step size
t_0 = 0  # We define the time at our initial observation as 0
t_N = 10
y_list,t_list = full_Euler (h,g,t_0,t_N,y_0)
plt.figure(figsize=(16,8))
plt.plot(t_list,y_list,label="Numerical",linewidth=1)
plt.plot(t_list,np.power(2,t_list),label="Analytical",linewidth=1)
plt.legend()
plt.grid()
plt.xlabel("days")
plt.ylable("bacteia")
plt.show()
```
